# Problem 1

```python
import numpy as np 
import matplotlib.pyplot as plt
import random 
%matplotlib inline
```

```python
n = 100
left = -1
right = 1
up = 2
down = -2
```

#the particle can either move: -x; +x; +y; or -y

```python
x = np.random.choice([left,right], size = n, p=[0.5,0.5,])
y = np.random.choice([up, down], size = n, p=[0.5,0.5,])
x_move = np.cumsum(x)
y_move= np.cumsum(y)
```

```python
plt.plot(x_move,y_move)
plt.xlim([-20, 20])
plt.ylim([-30, 30])
```

# Problem 2 

```python
s = np.random.choice([left,right], size=1000, p=[0.5, 0.5]).reshape(10,100)
bucket= np.sum(s,0)
```

summing over the zero axis gives the sum of 100 steps for 10 buckets = 1000 runs. yay. 

```python
plt.hist(bucket, bins = 10, density=True)
plt.xlim([-20, 20])
plt.xlabel('position x (total displacement)')
plt.ylabel('probability of x')
```

Huzzah! the plot is gaussian... which is cool becuase it represents a sort of average of a particle's position over time. There is a nice chance that we will find our particle at the origin.

# Problem 4

```python
from mpl_toolkits.mplot3d import Axes3D 
```

```python
run =1000
R = (np.random.rand(run)*6).astype("int") 
x = np.zeros(run) 
y = np.zeros(run)
z = np.zeros(run)
x[ R==0 ] = -1; x[ R==1 ] = 1 
y[ R==2 ] = -1; y[ R==3 ] = 1
z[ R==4 ] = -1; z[ R==5 ] = 1
x = np.cumsum(x) 
y = np.cumsum(y)
z = np.cumsum(z)
```

```python
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,) 
ax.scatter(x[-1],y[-1],z[-1])
plt.show()
```
